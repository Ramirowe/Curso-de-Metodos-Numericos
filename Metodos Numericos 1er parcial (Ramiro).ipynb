# Problema 2

Usando la notación del texto principal, el valor absoluto del error relativo se define como:

$$ |δx| = \left|\frac{\tilde{x} - x}{x}\right| $$

En la práctica, es conveniente proporcionar un límite para una cantidad distinta:

$$ |δ˜x| = \left|\frac{\tilde{x} - x}{\tilde{x}}\right| $$

Usando estas dos definiciones, intentemos encontrar una desigualdad que relacione $ |δ˜x| $ y $ |δx| $ (es decir, $ A \leq |δ˜x| \leq B $, donde $ A $ y $ B $ contienen $ |δx| $). 

Esta desigualdad implica ciertos límites sobre las magnitudes relativas de $ |δ˜x| $ y $ |δx| $ en situaciones prácticas.

Si tienes alguna pregunta o necesitas más aclaraciones, no dudes en preguntar.


Para resolver este problema y encontrar una desigualdad que relacione $|δ˜x|$ y $|δx|$, primero revisamos las definiciones dadas:

La definición de $|δx|$ es:

$$
|δx| = \left| \frac{\tilde{x} - x}{x} \right|
$$

Y la definición de $|δ˜x|$ es:

$$
|δ˜x| = \left| \frac{\tilde{x} - x}{\tilde{x}} \right|
$$

Para relacionar estas dos cantidades, observamos que ambas expresiones tienen la misma diferencia en el numerador $(x - \tilde{x})$ pero difieren en el denominador. Esto sugiere que la relación entre $|δx|$ y $|δ˜x|$ depende de la relación entre $x$ y $\tilde{x}$.

Las expresiones simplificadas son:

$$
|δx| = \left| 1 - \frac{\tilde{x}}{x} \right|
$$

$$
|δ˜x| = \left| 1 - \frac{x}{\tilde{x}} \right|
$$

Esto muestra que $|δx|$ y $|δ˜x|$ son formas inversas entre sí en términos de $x$ y $\tilde{x}$. Sin embargo, establecer una desigualdad directa entre $|δx|$ y $|δ˜x|$ que incluya límites específicos sin más información sobre la relación entre $x$ y $\tilde{x}$ es complejo.

Se entiende que $|δx|$ y $|δ˜x|$ ofrecen perspectivas diferentes sobre el error relativo y su magnitud relativa puede variar dependiendo de la relación entre $x$ y $\tilde{x}$. Aunque no se provee una desigualdad específica en forma de $A \leq |δ˜x| \leq B$, ambos errores relativos ofrecen una medida útil de la precisión de una aproximación, y su comparación directa requiere cuidado en la interpretación.

$$ P(x) = p_0 + p_1x + p_2x^2 + p_3x^3 + \ldots + p_{n-1}x^{n-1} \ \ \ (2.98) $$

Escribe una función que tome una lista que contiene los coeficientes $ p_i $, llamémosla `coeffs`, y el punto $ x $, y evalúe el valor de $ P(x) $ de la manera ingenua, es decir, de izquierda a derecha.

Observa que esta forma de codificar el polinomio contiene varias multiplicaciones innecesarias. Esto se debe a que $ x^i $ se evalúa como $ x \times x \times \ldots \times x $ (donde hay $ i - 1 $ multiplicaciones). Por lo tanto, esta forma de abordar el problema corresponde a:

$$ 1 + 2 + \ldots + n - 2 = \frac{(n - 1)(n - 2)}{2} \ \ \ (2.99) $$

multiplicaciones, desde $ x^2 $ hasta $ x^{n-1} $. Si reescribimos el polinomio:

$$ P(x) = p_0 + x(p_1 + x(p_2 + x(p_3 + \ldots + x(p_{n-2} + x p_{n-1}) \ldots))) \ \ \ (2.100) $$

entonces podemos hacerlo con solo $ n - 1 $ multiplicaciones (es decir, no se evalúan potencias). Esto es obviamente más eficiente, pero igualmente importante es el hecho de que de esta manera podemos limitar sustancialmente la acumulación de errores de redondeo (especialmente para polinomios de gran grado). Escribe una función que tome una lista que contiene los coeficientes $ p_i $, llamémosla `coeffs`, y el punto $ x $, y evalúe el valor de $ P(x) $ de la nueva manera, es decir, de derecha a izquierda.

Aplica las dos funciones anteriores al caso (aunque artificial) de:

```python
coeffs = [(-11)**i for i in reversed(range(8))]

````python
# Función para evaluar el polinomio de manera ingenua (de izquierda a derecha)
def evaluar_polinomio_ingenuo(coeffs, x):
    resultado = 0
    for i in range(len(coeffs)):
        resultado += coeffs[i] * (x ** i)
    return resultado


# Función para evaluar el polinomio de manera eficiente usando la regla de Horner (de derecha a izquierda)
def evaluar_polinomio_horner(coeffs, x):
    resultado = coeffs[-1]
    for i in range(len(coeffs) - 2, -1, -1):
        resultado = resultado * x + coeffs[i]
    return resultado

# Coeficientes del polinomio
coeffs = [(-11)**i for i in reversed(range(8))]

# Punto en el que evaluamos el polinomio
x = 11.01


# Imprimir resultados
print("Resultado utilizando evaluación ingenua:", evaluar_polinomio_ingenuo(coeffs, x))
print("Resultado utilizando la regla de Horner:", evaluar_polinomio_horner(coeffs, x))
````

# Problema 14

En compexp.py, empleamos la serie de Taylor para $e^x$. Haz lo mismo para $f(x) = \sin(x)$ en $x = 0.1$ y $x = 40$. Al igual que hicimos en el texto principal, primero necesitarás encontrar una relación simple entre el término n-ésimo y el término (n-1)-ésimo en la serie. Luego, usa una identidad trigonométrica para estudiar un $x$ equivalente (más pequeño) en lugar de 40.

````python
from math import factorial, sin, pi

# Implementación de la serie de Taylor para sin(x)
def compsin(x, terms=10):
    sum = 0.0
    for n in range(terms):  # Sumamos los primeros 'terms' términos de la serie
        term = ((-1)**n) * (x**(2*n + 1)) / factorial(2*n + 1)
        sum += term
    return sum

# Función para encontrar un x equivalente más pequeño para valores grandes de x
def find_smaller_x(x):
    # Utilizamos la identidad trigonométrica para simplificar x a un rango manejable
    return x % (2 * pi)

# Ejemplo de uso para x = 0.1 y x = 40
x_values = [0.1, 40]
for x in x_values:
    if x > 2*pi:  # Si x es mayor que 2*pi, buscamos un equivalente más pequeño
        x_equiv = find_smaller_x(x)
        print(f"x equivalente para {x}: {x_equiv}")
        result = compsin(x_equiv)
    else:
        result = compsin(x)
    print(f"sin({x}) ≈ {result}, valor real: {sin(x)}")
````

## a) Relación de Recurrencia de Bonnet Generalizada

La generalización de la relación de recurrencia de Bonnet, Ec. (2.86), para polinomios de Legendre asociados, toma la forma:

$$(n - m)P^n_m(x) = x(2n - 1)P^{n-1}_m(x) - (n + m - 1)P^{n-2}_m(x) \ \ \ (2.125)$$
Para comenzar, apreciarás tener acceso a:

$$P^m_m(x) = (-1)^m(2m - 1)!!(1 - x^2)^{\\frac{m}{2}} \ \ \ (2.126)$$

lo cual hace uso del factorial doble (el producto de enteros impares). Otra expresión de inicio útil es:

$$P^{m+1}_m(x) = x(2m + 1)P^m_m(x) \ \ \ (2.127)$$

Calcula $P^n_m(x)$ avanzando a través de la Ec. (2.125), haciendo uso de la Ec. (2.126) y la Ec. (2.127) según sea necesario. Utiliza tu función para graficar $P^4_2(x)$.

El enfoque anterior funciona para $0 \leq m \leq n $. Para tratar con valores de $m$ negativos, usa:

$$P^n_{-m}(x) = (-1)^m \\frac{(n - m)!}{(n + m)!} P^n_m(x) \ \ \ (2.128)$$

Grafica $P^4_{-2}(x)$; si quieres, puedes compararlo contra `scipy.special.lpmv()`.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial2, gamma
from mpl_toolkits.mplot3d import Axes3D

# Implementación de la relación de recurrencia de Bonnet generalizada para Pnm
def Pnm(n, m, x):
    if m < 0:
        return ((-1)**m) * gamma(n-m+1) / gamma(n+m+1) * Pnm(n, -m, x)
    if m == n:
        return (-1)**m * factorial2(2*m - 1) * (1 - x**2)**(m/2)
    if n == m + 1:
        return x * (2*m + 1) * Pnm(m, m, x)
    return ((x * (2*n - 1) * Pnm(n-1, m, x) - (n + m - 1) * Pnm(n-2, m, x)) / (n - m))

# Calculando y graficando P^4_{-2}(x)
x_values = np.linspace(-1, 1, 400)
P4_neg2_values = np.array([Pnm(4, -2, x) for x in x_values])

plt.figure(figsize=(12, 6))
plt.plot(x_values, P4_neg2_values, label='$P^4_{-2}(x)$')
plt.xlabel('$x$')
plt.ylabel('$P^4_{-2}(x)$')
plt.legend()
plt.title('Polinomio de Legendre Asociado $P^4_{-2}(x)$')
plt.grid(True)
plt.show()
````

  (-1)^m \sqrt{\frac{2n+1}{4\pi} \frac{(n-m)!}{(n+m)!}} P_n^m(\cos \theta) \sin(m\phi), & \text{if } m < 0 \\
  \sqrt{\frac{2n+1}{4\pi}} P_n^0(\cos \theta), & \text{if } m = 0 \\
  (-1)^m \sqrt{\frac{2n+1}{4\pi} \frac{(n-m)!}{(n+m)!}} P_n^m(\cos \theta) \cos(m\phi), & \text{if } m > 0 
\end{cases}
$$

````python
# Función para calcular los armónicos esféricos reales Ynm
def Ynm(n, m, theta, phi):
    P = Pnm(n, abs(m), np.cos(theta))
    prefactor = np.sqrt((2*n + 1) / (4*np.pi) * gamma(n-abs(m)+1) / gamma(n+abs(m)+1))
    
    if m > 0:
        return (-1)**m * prefactor * P * np.cos(m * phi)
    elif m == 0:
        return prefactor * P
    else:
        return (-1)**m * prefactor * P * np.sin(abs(m) * phi)

# Creación de la cuadrícula para theta y phi para visualización
theta, phi = np.meshgrid(np.linspace(0, np.pi, 100), np.linspace(0, 2*np.pi, 100))

# Cálculo de |Y42|
Y42 = np.abs(Ynm(4, 2, theta, phi))

# Convertir a coordenadas cartesianas para la visualización
x = Y42 * np.sin(theta) * np.cos(phi)
y = Y42 * np.sin(theta) * np.sin(phi)
z = Y42 * np.cos(theta)
````
